```python
import string
idx2vocab = dict((idx, token) for idx, token in enumerate(vocab))
vocab2idx = dict((v, k) for k, v in idx2vocab.items())
idx2ch = dict((idx, ch) for idx, ch in enumerate(string.printable[:len(idx2vocab)]))
ch2idx = dict((v, k) for k, v in idx2ch.items())
```

```python
def t2cht(template):
    tokens = basic_program_normalize(template)
#     print(tokens)
    indices = [vocab2idx[token] for token in tokens]
#     print(indices)    
    cht = [idx2ch[idx] for idx in indices]
    return "".join(cht)

def cht2t(cht):
    indices = [ch2idx[ch] for ch in cht]
#     print(indices)    
    template = [idx2vocab[idx] for idx in indices]
    return " ".join(template)
```

```python
def delete_redundancy(grammar):
    terminals = {}
    for k, v in grammar.items():
        flag = True
        for vv in v:
            if not isinstance(vv, str):
                flag = False
                break
        if flag:
            terminals.update({k:" ".join(v)})
    
    for k in terminals.keys():
        del grammar[k]
    
    grammar_ = {}
    for k, v in grammar.items():
        grammar_[k] = [terminals.get(vv, vv) for vv in v]
    return grammar_
```

```python
import re

def basic_program_normalize(line):
    _patterns = [r'\'',
                 r'\"',
                 r'\.',
                 r'<br \/>',
                 r',',
                 r'\(',
                 r'\)',
                 r'\!',
                 r'\?',
                 r'\;',
                 r'\:',
                 r'\s+']
    _replacements = [' \' ',
                     ' \" ',
                     ' . ',
                     ' ',
                     ' , ',
                     ' ( ',
                     ' ) ',
                     ' ! ',
                     ' ? ',
                     ' \; ',
                     ' \: ',
                     ' ']
    _patterns_dict = list((re.compile(p), r) for p, r in zip(_patterns, _replacements))
    for pattern_re, replaced_str in _patterns_dict:
        line = pattern_re.sub(replaced_str, line)
    return line.split()
```
